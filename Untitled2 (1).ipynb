```python
numerical_columns = barcelona_airbnb.select_dtypes(include=['int', 'float']).columns
numerical_columns
```

```python
for i in numerical_columns:
  plt.hist(barcelona_airbnb[i], bins=20, edgecolor='black')

  # Adding labels and title
  plt.xlabel(i)
  plt.ylabel('Frequency')
  plt.title(f'Histogram of {i}')

  # Display the histogram
  plt.show()
```

```python
# outliers check

for i in numerical_columns:
  plt.boxplot(barcelona_airbnb[i])

  # Adding labels and title
  plt.xlabel(i)
  plt.ylabel('Frequency')
  plt.title(f'boxplot of {i}')

  # Display the histogram
  plt.show()
```

```python
for i in numerical_columns:
  plt.figure(figsize=(8, 6))
  plt.scatter(barcelona_airbnb[i], barcelona_airbnb['price'], c='blue', marker='o')

  # Adding labels and title
  plt.xlabel(i)
  plt.ylabel('price')
  plt.title(f'Relation between the {i} and price')

  # Display the scatter plot
  plt.show()
```
